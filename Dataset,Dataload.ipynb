## 미니 배치와 데이터 로드

### 미니 배치와 배치 크기  
- 미니배치만큼 가져가 미니 배치에 대한 비용을 계산하고 경사 하강법 수행  
- 미니배치가 다 돌아서 전체 데이터에 대한 학습이 1회 끝나면 1회 epoch가 끝난다  
- 미니배치 = batch_size

```python
x_train = torch.FloatTensor([[73, 80, 75],
                             [93, 88, 93],
                             [89, 91, 90],
                             [96, 98, 100],
                             [73, 66, 70]])
y_train = torch.FloatTensor([[152], [185], [180], [196], [142]])
```

### 데이터 로드하기

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
from torch.utils.data import TensorDataset # 텐서 데이터셋
from torch.utils.data import DataLoader # 데이터 로더
```

```python
dataset=TensorDataset(x_train,y_train)
```

```python
dataloader=DataLoader(dataset,batch_size=2, shuffle=True)
```

```python
class Multi(nn.Module):
  def __init__(self):
    super().__init__()
    self.linear = nn.Linear(3,1)

  def forward(self,x):
    return self.linear(x)

model=Multi()

optimizer=torch.optim.SGD(model.parameters(),lr=1e-5)
```

```python
epochs=200

for epoch in range(epochs+1):
  for batch_idx, samples in enumerate(dataloader):
    #print(batch_idx)
    #print(samples)

    x_train,y_train = samples

    prediction=model(x_train)

    cost=F.mse_loss(prediction,y_train)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()


    print('Epoch {:4d}/{} Batch {}/{} Cost: {:.6f}'.format(
        epoch, epochs, batch_idx+1, len(dataloader),
        cost.item()
        ))
```

```python
new_var =  torch.FloatTensor([[73, 80, 75]])
pred_y = model(new_var)
print("훈련 후 입력이 73, 80, 75일 때의 예측값 :", pred_y)
```

### 커스텀 데이터셋으로 선형회귀 구현

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
class Multi(nn.Module):
  def __init__(self):
    super().__init__()
    self.linear=nn.Linear(3,1)

  def forward(self,x):
    return self.linear(x)

model= Multi()
optimizer=torch.optim.SGD(model.parameters(),lr=1e-5)
```

```python
class CustomDataset(torch.utils.data.Dataset): # 데이터셋 상속

  def __init__(self):
    self.x_train=[[73, 80, 75],
                   [93, 88, 93],
                   [89, 91, 90],
                   [96, 98, 100],
                   [73, 66, 70]]
    self.y_train=[[152], [185], [180], [196], [142]]

  def __len__(self): # 총 데이터의 개수를 리턴
    return len(self.x_train)

  def __getitem__(self,idx): # 인덱스를 입력받아 그에 맵핑되는 입출력 데이터를 파이토치의 Tensor형태로 리턴
    x= torch.FloatTensor(self.x_train[idx])
    y= torch.FloatTensor(self.y_train[idx])
    return x,y
```

```python
dataset=CustomDataset()
dataloader= DataLoader(dataset,batch_size=2, shuffle=True)
```

```python
len(dataset)
```

```python
epochs=200
for epoch in range(epochs+1):
  for idx,samples in enumerate(dataloader):
    x_train,y_train=samples

    prediction=model(x_train)

    cost=F.mse_loss(prediction, y_train)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    print('Epoch {:4d}/{} Batch {}/{} Cost: {:.6f}'.format(
        epoch,epochs, idx+1, len(dataloader),
        cost.item()
        ))
```
